```python
import pandas as pd

## read in data
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

## head
display(train.head())

# describe columns
display(train.select_dtypes('object').describe().T)

display(train.describe().T.round(2))
```

```python
## revenue min and max
print('min revenue = ' , min(train['revenue']))
print('max revenue = ' , max(train['revenue']))
print('number of null values = ' , train['revenue'].isna().sum())

## plot revenue distribution
train['revenue'].plot.hist(bins=50, range = (0, 100000000))
```

```python
## popularity min and max
print('min popularity = ' , min(train['popularity']))
print('max popularity = ' , max(train['popularity']))
print('number of null values = ' , train['popularity'].isna().sum())

## plot popularity distribution
train['popularity'].plot.hist(bins=50, range = (0, 30))
```

```python
## xyplot
import matplotlib.pyplot as plt
import numpy as np

b, m = np.polyfit(train['popularity'], train['revenue'], 1)

plt.plot(train['popularity'], train['revenue'], '.')
plt.plot(train['popularity'], b + m * train['popularity'], '-')

plt.xlim(0, 50)
plt.ylim(0, 1000000000)
plt.ylabel('Revenue')
plt.xlabel('Popularity')
```

```python
## popularity min and max
print('min runtime = ' , min(train['runtime']))
print('max runtime = ' , max(train['runtime']))
print('number of null values = ' , train['runtime'].isna().sum())

## replace the missing values with mean
train['runtime'].fillna(value=numpy.mean(train['runtime']), axis=0, inplace=True)

## plot popularity distribution
train['runtime'].plot.hist(bins=50, range = (0, 350))
```
